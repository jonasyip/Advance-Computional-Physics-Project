```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import matplotlib.ticker as tck
```

```python
static = pd.read_csv("guided.csv", header=0)
threads = static["threads"]
execution_time = static["execution_time_(s)"]
```

```python
threads_plot = []
guided_execution_time_avg = []
guided_error = []
guided_speed_up = []
guided_efficiency = []

T_1 = (execution_time[0] + execution_time[1] + execution_time[2]) / 3
error = (max([execution_time[0], execution_time[1], execution_time[2]]) - min([execution_time[0], execution_time[1], execution_time[2]])) / 4
guided_error.append(error)
guided_efficiency.append(1)
threads_plot.append(2)
guided_speed_up.append(2)

for i in range(3, len(threads), 3):
    avg = (execution_time[i] + execution_time[i+1] + execution_time[i+2]) / 3
    S_p = T_1 / avg
    guided_speed_up.append(S_p)
    Eff_p = S_p / threads[i]
    guided_efficiency.append(Eff_p)
    
    error = np.std([execution_time[i], execution_time[i+1], execution_time[i+2]]) / np.sqrt(3)
    guided_error.append(error)

    threads_plot.append(threads[i])
```

```python
static = pd.read_csv("dynamic.csv", header=0)
threads = static["threads"]
execution_time = static["execution_time_(s)"]
```

```python
threads_plot = []
dynamic_execution_time_avg = []
dynamic_error = []
dynamic_speed_up = []
dynamic_efficiency = []

T_1 = (execution_time[0] + execution_time[1] + execution_time[2]) / 3
error = (max([execution_time[0], execution_time[1], execution_time[2]]) - min([execution_time[0], execution_time[1], execution_time[2]])) / 4
dynamic_error.append(error)
dynamic_efficiency.append(1)
dynamic_speed_up.append(2)
threads_plot.append(2)

for i in range(3, len(threads), 3):
    avg = (execution_time[i] + execution_time[i+1] + execution_time[i+2]) / 3
    S_p = T_1 / avg
    dynamic_speed_up.append(S_p)
    Eff_p = S_p / threads[i]
    dynamic_efficiency.append(Eff_p)
    
    error = np.std([execution_time[i], execution_time[i+1], execution_time[i+2]]) / np.sqrt(3)
    dynamic_error.append(error)

    threads_plot.append(threads[i])
```

```python
xticks = [x for x in range(0, 30, 2)]


fig, ax = plt.subplots(1,figsize=(8,6))

#plt.errorbar(threads_plot, guided_efficiency, guided_error, fmt='.', color='Black', ecolor='LightGrey')
plt.scatter(threads_plot, static_efficiency, label="Static", color="blue")
plt.plot(threads_plot, static_efficiency, color="blue", alpha=0.5)

plt.scatter(threads_plot, dynamic_efficiency, label="Dynamic", color="orange")
plt.plot(threads_plot, dynamic_efficiency, color="orange", alpha=0.5)

plt.scatter(threads_plot, guided_efficiency, label="Guided", color="green")
plt.plot(threads_plot, guided_efficiency, color="green", alpha=0.5)

plt.xticks(xticks, xticks)

plt.xlabel("Threads", size=16)
plt.ylabel("Efficiency", size=16)

ax.xaxis.set_tick_params(labelsize=16)
ax.yaxis.set_tick_params(labelsize=16)

ax.xaxis.set_minor_locator(tck.AutoMinorLocator(2))
ax.yaxis.set_minor_locator(tck.AutoMinorLocator())

ax.legend(prop={'size': 17})
plt.show()
```

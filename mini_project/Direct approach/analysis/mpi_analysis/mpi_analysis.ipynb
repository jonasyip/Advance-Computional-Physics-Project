```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import matplotlib.ticker as tck
```

```python
static = pd.read_csv("mpi_threads_500.csv", header=0)
processes = static["processes"]
execution_time = static["execution_time_(s)"]
```

```python
processes_plot = []
mpi500_execution_time_avg = []
mpi500_error = []
mpi500_speed_up = []
mpi500_efficiency = []

T_1 = (execution_time[0] + execution_time[1] + execution_time[2]) / 3
mpi500_execution_time_avg.append(T_1)
error = (max([execution_time[0], execution_time[1], execution_time[2]]) - min([execution_time[0], execution_time[1], execution_time[2]])) / 2
mpi500_error.append(error)
mpi500_efficiency.append(1)
processes_plot.append(2)
mpi500_speed_up.append(2)

for i in range(3, len(processes), 3):
    avg = (execution_time[i] + execution_time[i+1] + execution_time[i+2]) / 3
    mpi500_execution_time_avg.append(avg)
    S_p = T_1 / avg
    mpi500_speed_up.append(S_p)
    Eff_p = S_p / processes[i]
    mpi500_efficiency.append(Eff_p)
    
    error = (max([execution_time[i], execution_time[i+1], execution_time[i+2]]) - min([execution_time[i], execution_time[i+1], execution_time[i+2]])) / 2
    mpi500_error.append(error)

    processes_plot.append(processes[i])
```

```python
xticks = [x for x in range(0, 30, 2)]


fig, ax = plt.subplots(1,figsize=(8,6))

ax.errorbar(processes_plot, mpi1000_execution_time_avg, mpi1000_error, fmt=".",label="N = 1000", color="blue", ecolor="lightblue")
ax.plot(processes_plot, mpi1000_execution_time_avg, color="blue", alpha=0.5)

ax.errorbar(processes_plot, mpi500_execution_time_avg, mpi500_error, fmt=".",label="N = 500", color="green", ecolor="lightgreen")
ax.plot(processes_plot, mpi500_execution_time_avg, color="green", alpha=0.5)

ax.errorbar(processes_plot, mpi100_execution_time_avg, mpi500_error, fmt=".",label="N = 100", color="orange", ecolor="wheat")
ax.plot(processes_plot, mpi100_execution_time_avg, color="orange", alpha=0.5)

ax.set_yscale('log',base=10) 
plt.xticks(xticks, xticks)

ax.set_xlabel("Processes", size=16)
ax.set_ylabel("Time (s)", size=16)


ax2=ax.twinx()

ax2.plot(processes_plot, mpi1000_speed_up, color="green",linestyle="--")

ax2.plot(processes_plot, mpi500_speed_up, color="blue",linestyle="--")

ax2.plot(processes_plot, mpi100_speed_up, color="orange",linestyle="--")


ax.xaxis.set_tick_params(labelsize=16)
ax.yaxis.set_tick_params(labelsize=16)

ax.xaxis.set_minor_locator(tck.AutoMinorLocator(2))
ax.tick_params(which='minor', length=2.5)

ax.yaxis.set_major_formatter(tck.FuncFormatter(lambda y,pos: ('{{:.{:1d}f}}'.format(int(np.maximum(-np.log10(y),0)))).format(y)))



ax2.set_ylabel("Speedup", size=16)
ax2.yaxis.set_tick_params(labelsize=16)

ax2.yaxis.set_minor_locator(tck.AutoMinorLocator())

ax.legend(prop={'size': 14})
plt.show()
```

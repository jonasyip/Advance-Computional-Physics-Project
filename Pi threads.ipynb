```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
threads = [1, 2, 4, 8, 16, 20, 24, 28]
time_taken = [0.116560, 0.065745, 0.038069, 0.020629, 0.011065, 0.008889, 0.007706, 0.006770]
```

```python
#Blue Crystal 4
plt.title("Threads vs time taken to process Pi in 10000000 iterations")
plt.plot(threads, time_taken, alpha=0.5)
plt.scatter(threads, time_taken)
plt.ylabel("Time taken (seconds)")
plt.xlabel("Number of threads")
plt.show()
```
